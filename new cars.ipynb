```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
from datetime import datetime
```

```python
# https://fred.stlouisfed.org/series/CUUR0000SETA01
new_cars=pd.read_csv("Data_files/new_cars_trucks.csv", index_col=0, parse_dates=True)
new_cars
```

```python
value_new_cars = new_cars['Value']
value_new_cars
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

sns.lineplot(value_new_cars)

crisis_data = [
    (datetime(2020, 3, 1), "Start of coronavirus")
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, value_new_cars.asof(date) + 1),
                xytext=(date, value_new_cars.asof(date) + 5),
                arrowprops=dict(facecolor="black", headwidth=4, width=.6,
                                headlength=2),
                horizontalalignment="left", verticalalignment="top")

ax.set_title("\nValue of new motor vehicles\n", fontsize=25)
ax.set_xlabel("\nDate", fontsize=20)
ax.set_ylabel("Value", fontsize=20)
ax.legend(["New cars and trucks"], title="Category", edgecolor="#1c1c1c")
sns.despine(left=True, bottom=True)

plt.show()
```
